# GRAFICA DE LA FUNCION SENO.

```python
import matplotlib.pyplot as plt
from numpy import arange,sin,pi,cos,tan
```

```python
#Definimos variable que dara los valores de x
x = arange(0.0,4*pi,0.01)
```

```python
y = sin(x)
```

```python
plt.plot(x,y) #Con esta funcion graficamos x & y.
plt.grid(True) #Con esta funcion mostramos la cuadricula.
plt.ylabel('Y') #Con esta funcion pintamos la etiqueta Y
plt.ylabel('X') #Con esta funcion pintamo la etiqueta X
plt.show() #Con esta funcion mostramos la grafica
```

# GRAFICA DE LA FUNCION COSENO

```python
b = arange(0.0,4*pi,0.01)
d = cos(b)
```

```python
plt.plot(b,d) #Con esta funcion graficamos x & y.
plt.grid(True) #Con esta funcion mostramos la cuadricula.
plt.ylabel('b') #Con esta funcion pintamos la etiqueta Y
plt.ylabel('d') #Con esta funcion pintamo la etiqueta X
plt.show() #Con esta funcion mostramos la grafica
```

# GRAFICA DE LA FUNCION TANGENTE.

```python
f = arange(0.0,4*pi,0.01)
m = tan(f)
```

```python
plt.plot(f,m) #Con esta funcion graficamos x & y.
plt.grid(True) #Con esta funcion mostramos la cuadricula.
plt.ylabel('f') #Con esta funcion pintamos la etiqueta Y
plt.ylabel('m') #Con esta funcion pintamo la etiqueta X
plt.show() #Con esta funcion mostramos la grafica
```
